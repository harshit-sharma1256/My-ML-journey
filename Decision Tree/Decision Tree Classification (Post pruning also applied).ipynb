```python
x= data.iloc[:,:-1]
y= data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.33,random_state = 42)
```

## Fitting the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

## Now constructing the tree

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(model,filled=True)
#CART Tree(Binary Split)
plt.show()
```

###### The Above tree signifies:-
###### 1. X[2] <= 2.5 , means feature X[2] == "petal_length" is taken for split
###### 2. Gini impurity is used to check the purity of split . if gini ==  0 ----> pure split else impure split
###### 3. Total samples in this selected feature.
###### 4. Value = [31,35,34] -----> [31 setosa , 35 verginica , 34 versicolor]

# See at level 2 (level starts from 0) .
# In green box there are 34 verginica out of total 38 samples, rest 4 are versicolor.
# In purple box there are 30 versicolor out of total 31 sampples, rest 1 is verginica.

# So there is a high ratio gap in these boxes. We do not need to grow the DT till max_depth .
# I will fix the max_depth  == 2 in order to reduce overfitting (Post Pruning).


```python
# Post pruning
model1 = DecisionTreeClassifier(max_depth =2)
model1.fit(x_train,y_train)
```

```python
plt.figure(figsize = (15,10))
tree.plot_tree(model1,filled=True)
#CART Tree(Binary Split)
plt.show()
```

#### Above is the tree till max_depth ==2 . Post pruning has been performed on this Decision Tree.

#### Now if a test point goes to green node , answer is 'verginica'. 
#### Now if a test point goes to purple node , answer is 'versicolor'.

```python
# We can also play with other parameters in Post pruning.
# But i will use them in the Pre-Pruning.
```

```python
y_pred = model1.predict(x_test)
```
